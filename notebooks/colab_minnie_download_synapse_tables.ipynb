```python
from caveclient import CAVEclient
import pandas as pd
import numpy as np
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
drive_path = '/content/drive/MyDrive/minnie65/synapse_tables/'
```

```python
from google.colab import userdata

os.environ["CAVECLIENT_TOKEN"] = userdata.get('CAVECLIENT_TOKEN')
caveclient_key = os.getenv('CAVECLIENT_TOKEN')
#print(caveclient_key)
```

```python
# caveclient_key
```

```python
client = CAVEclient("minnie65_public", auth_token=caveclient_key)
```

```python
client.materialize.get_versions()
```

```python
for version in client.materialize.get_versions():
    print(f"Version {version}: {client.materialize.get_timestamp(version)}")
```

```python
client.version=1181
```

```python
my_root_id = 864691135776009133 #864691135587226354 #864691136237449532 #864691135866387734 #864691135675330756 #864691136781187182
```

```python
# Pull output synapses for my_root_id
syn_df_pre = client.materialize.synapse_query(pre_ids=my_root_id)
print(f"Total number of output synapses for {my_root_id}: {len(syn_df_pre)}")
syn_df_pre.head()
```

```python
# Save the DataFrame as a CSV file
syn_df_pre.to_csv(f"{drive_path}/{my_root_id}_syn_df_pre.csv", index=False)
```

```python
# pull input synapses for my_root_id
syn_df_post = client.materialize.synapse_query(post_ids=my_root_id)
print(f"Total number of input synapses for {my_root_id}: {len(syn_df_post)}")
syn_df_post.head()
```

```python
# Save the DataFrame as a CSV file
syn_df_post.to_csv(f"{drive_path}/{my_root_id}_syn_df_post.csv", index=False)
```
